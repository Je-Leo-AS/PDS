## 1) Gerar os seguintes sinais e visualiza-los (comandos plot e stem). 

```python
from pylab import *
```

### a) impulso unitário, para valores de n (abscissa) entre -10 e 10 (stem), inclusive.

```python
n = np.arange(-10,11,1)
impulso = (n==0)*1
plt.stem(n,impulso)
plt.show()
```

### b) degrau unitário, para n entre -10 e 10 (stem), inclusive.


```python
degrau = (n>=0)*1
plt.stem(n,degrau)
plt.show()
```

### c) ruido branco gaussiano (média zero, variância unitária) com 40 valores, para n=0 a 39 (stem).

```python
amostras = 40
n = np.arange(0,amostras,1)
ruido = np.random.normal(0, 1, amostras)
plt.stem(n,ruido)
plt.show()
```

### d) senoide com 4 ciclos, com 90 valores, para n=0 a 89 (stem).

```python
n = np.arange(0,90,1)
y = np.linspace(0, 4*2*np.pi, 90)
seno = np.sin(y)
plt.stem(n,seno)
plt.show()
```

```python
n = np.arange(0,90,1)
y = np.linspace(0, 4*2*np.pi, 90)
cosseno = np.cos(y)
plt.stem(n,cosseno)
plt.show()
```

### e) cosenoide com 5 ciclos, com 110 valores, para n=0 a 109 (stem). 

```python
amostras= 110
n = np.arange(0,amostras,1)
y = np.linspace(0, 5*2*np.pi, amostras)
cosseno = np.cos(y)
plt.stem(n,cosseno)
plt.show()
```

### f) exponencial complexa com 5 ciclos, para n=0 a 69. Parte real (stem) e parte imaginária (plot). Usar subplot para plotar em dois graficos separados: subplot(2,1,1) para o primeiro e subplot(2,1,2) para do segundo.

```python
amostras= 70
n = np.arange(0,amostras,1)

exponential_complex = np.exp(1j * 2 * np.pi * 5 * n / amostras)
real_part = np.real(exponential_complex)
imaginary_part = np.imag(exponential_complex)

plt.subplot(2, 1, 1)
plt.stem(n, real_part, basefmt='b', linefmt='b-', markerfmt='bo')

plt.subplot(2, 1, 2)
plt.plot(n, imaginary_part, 'g-')

plt.tight_layout()
plt.show()
```

### g) senoide com frequência w=0,2 rd ( sin(w*n) ) , para n de 0 a 99 (plot).

```python
amostras = 100
n = np.arange(0, amostras)
frequency = 0.2

sine_wave = np.sin( frequency * n)

plt.plot(n, sine_wave, 'g-')

plt.show()
```

### h) senoide de 60 Hz, (amostragem = 800 amostras/s), com 70 valores, para n de 0 a 69 (plot).

```python
amostras = 70
amostragem = 800
n = np.arange(0, amostras)
frequency = 60
periodo = 1/frequency
intervalo = 1 / amostragem
sine_wave = np.sin(2 * np.pi * frequency * n * intervalo)

plt.plot(n, sine_wave)

plt.show()
```

### l) exponencial decrescente para n = 0 a 60 (stem).

```python
amostras = 60
n = np.arange(0, amostras)
frequency = 1e-1

expo = 1 * np.exp(- frequency * n)

plt.stem(n, expo)

plt.show()
```

### m) exponencial oscilante decrescente $x[n]=z^n$, para n=0 a 90, parte real (stem).

```python
amostras = 91
n = np.arange(0, amostras)

amplitude = 1.027
frequency = 50.57
expo = amplitude*np.exp(1j*frequency)

z = expo**(-1*n)
real = z.real
a = stem(n,real)

plt.show()
```

### n) exponencial oscilante crescente $x[n]=z^n$, para n=0 a 80, parte imaginaria (stem).

```python
amostras = 91
n = np.arange(0, amostras)

amplitude = 1.027
frequency = 50.57
expo = amplitude*np.exp(1j*frequency)

z = expo**(-1*n)
imag = z.imag
a = stem(n,imag)

plt.show()
```
