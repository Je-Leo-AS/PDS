```python
# a)
N=100
x1=concatenate((zeros(10),ones(10),zeros(10)))

X1t=tfourier(x1,N)
X1f=fft(x1)
```

```python
figure(figsize=(15,4))
subplot(1,2,1)
plot(x1,label='x1')
xlabel('n')
legend()
subplot(1,2,2)
xlabel('w (rad)')

k=2*pi*arange(N)/N
kk=2*pi*arange(len(X1f))/len(X1f)
plot(k, abs(X1t),label='tfourier')
plot(kk, abs(X1f),'o',label='fft')
legend()
show()
```
