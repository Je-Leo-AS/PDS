1) Considere a sequência teórica: $x[n]=1$, para $n=0,1,2,3$, e $x[n]=0$, para todos os outros valores de $n$ ($-\infty ... +\infty)$.

2) Calcule a transformada de fourier da sequencia $x[n]$, $n=0$ a 199, obtendo 200 valores na faixa de frequência entre 0 e $2\pi$.


$ X(f) = \sum_{n=0}^{N-1} x[n] \cdot e^{-j2\pi fn/N}  $

- \(X(f)\) é a Transformada de Fourier da sequência.
- \(x[n]\) é a sequência de entrada no domínio do tempo.
- \(N\) é o número de amostras na sequência.
- \(f\) é a frequência em termos de fração de ciclos por amostra 


```python
N = 200
x = np.zeros(N)
x[:4] = 1
F1 = linspace(0,2*np.pi,N)
X1 = 0 
for i in range(N):
    X1 = X1 + x[i]*exp(-1j*F1*i)

plot(F1, abs(X1))
```

3) Calcule a DFT desta sequencia, para n=0 a 3.

```python
N = 4
x = np.zeros(N)
x[:4] = 1
F2 = linspace(0,2*np.pi,N, endpoint=False)
X2 = 0 
for i in range(N):
    X2 = X2 + x[i]*exp(-1j*F2*i)

plot(F2, abs(X2))
```

4) Calcule a DFT desta sequencia, para n=0 a 7.

```python
N = 8
x = np.zeros(N)
x[:4] = 1
F3 = linspace(0,2*np.pi,N, endpoint=False)
X3 = 0 
for i in range(N):
    X3 = X3 + x[i]*exp(-1j*F3*i)

plot(F3, abs(X3))
```

5) Calcule a DFT desta sequencia, para n=0 a 15.

```python
N = 16
x = np.zeros(N)
x[:4] = 1
F4 = linspace(0,2*np.pi,N, endpoint=False)
X4 = 0 
for i in range(N):
    X4 = X4 + x[i]*exp(-1j*F4*i)

plot(F4, abs(X4))
```

6) Plote no mesmo eixo (abscissa de zero a 2$\pi$), todas as 4 transformadas. usando os seguintes simbolos: 2) linha continua, 3) 'o', 4) '+', 5) 'x' .

```python
plot(F1, abs(X1), label='200')
plot(F2, abs(X2), 'o', label='o')
plot(F3, abs(X3), '+', label='+')
plot(F4, abs(X4), 'x', label='x')
legend()
```

É possível observar o efeito da aplicação de diferentes janelas retangulares ao calcular a Transformada Discreta de Fourier (DFT) de um sinal. Além disso, podemos analisar o impacto no resultado da DFT ao introduzir zeros após o sinal.

O sinal original consistia em um pulso retangular, cuja transformada é representada pela função sinc (sin(x)/x). Isso é claramente visível na linha contínua (X2). No entanto, com apenas 4 pontos (X3), o pulso retangular parece se assemelhar a um sinal constante. De fato, a transformada X3 exibe apenas um pico na frequência zero, com valores zero em todas as outras frequências. À medida que preenchemos esse sinal com zeros até 8 pontos (X4), começamos a discernir mais pontos da função sinc esperada. O mesmo ocorre quando aumentamos para 16 pontos (X5).
